```python
from langchain_community.llms import Replicate
import os
from google.colab import userdata
# Set the API token
api_token = userdata.get('REPLICATE_API_TOKEN')
os.environ["REPLICATE_API_TOKEN"] = api_token
# Model setup
model = "ibm-granite/granite-3.3-8b-instruct"
output = Replicate(
model=model,
replicate_api_token=api_token,
)
```

```python
import pandas as pd

# Load the JSON dataset
try:
    # Replace 'your_dataset.json' with the name of your JSON file
    df = pd.read_json('csvjson.json')
    print("JSON dataset loaded successfully!")
    display(df.head()) # Display the first few rows to verify
except FileNotFoundError:
    print("Error: JSON file not found. Make sure the file name and path are correct.")
except Exception as e:
    print(f"An error occurred during data loading: {e}")
```

```python
# Refine top_k model parameter value
kualifikasi = {
"top_k": 3,
"top_p": 1.0,
"max_tokens": 3000,
"min_tokens": 0,
"random_seed": None,
"repetition_penalty": 1.0,
"stopping_criteria": "length (2000 tokens)",
"stopping_sequence": None
}


if 'df' in locals() and not df.empty:
    refined_prompt = f"""
    Identify patterns and trends in this stock market data that could inform investment strategies by
    Stocks showing consistent growth patterns, Undervalued stocks based on PE ratios and market metrics,
    High-dividend yield opportunities,Volatility patterns that might indicate market timing opportunities,
    Provide actionable investment insights based on the data patterns:
    {df.to_string()}
    """

    # Invoke the model with the example prompt
    response = output.invoke(refined_prompt,parameters = kualifikasi)

    # Print the response
    print("\nGranite Model Refined Response:\n")
    print(response)
else:
    print("DataFrame is empty or not loaded. Cannot proceed with the prompt.")
```
